```python
# Importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Importation du jeu de données
import pandas as pd
data = pd.read_csv('data.csv', encoding = 'latin-1')
data
```

```python
#Liste des differentes colonnes
data.columns
```

```python
#Description du jeu de données
data.describe()
```

```python
# suppression de la derniere colonne
del data['Unnamed: 32']
#Suppression des doublons
data.drop_duplicates(inplace = True)
```

```python
#Description du jeu de données
data.describe()
```

```python
# X = data.iloc[:, 2:]   pour avoir le dataset des features
#pour le reseau de neurones ce sont les valeurs de la dataset des features qui nous interesse 
#Recuperer les valeurs des features 
X = data.iloc[:, 2:].values
X = pd.DataFrame(X)
X
```

```python
#Recuperer la target
#y = data.iloc[:, 1]
#Recuperer les valeurs de la target
y = data.iloc[:, 1].values
y = pd.DataFrame(y)
y
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder
labelencoder_X_1 = LabelEncoder()
y = labelencoder_X_1.fit_transform(y)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 0)
```

```python
ligne, colonne = X_train.shape
#print(X_train.shape[0])
print("ligne : ", ligne, "\ncolonne : ", colonne)
```

```python
ligne, colonne = X_test.shape
#print(X_test.shape[0])
print("ligne : ", ligne, "\ncolonne : ", colonne)
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
import tensorflow as tf
```

```python
# Initialising the ANN(Artificial Neural Netword)
classifier = Sequential()
```

```python
#deux couches cachees de 16 neurones par couche cachee et un neuron pour la couche de sortie
sizes = [16, 16, 1]
```

```python
# Models 
# premiere couche cachee
Momentum = Sequential([Dense(units=sizes[0], input_dim=X_train.shape[1], activation='relu')])
Nag = Sequential([Dense(units=sizes[0], input_dim=X_train.shape[1], activation='relu')])
Adagrad = Sequential([Dense(units=sizes[0], input_dim=X_train.shape[1], activation='relu')])
Adadelta = Sequential([Dense(units=sizes[0], input_dim=X_train.shape[1], activation='relu')])
RMSProp = Sequential([Dense(units=sizes[0], input_dim=X_train.shape[1], activation='relu')])
Adam = Sequential([Dense(units=sizes[0], input_dim=X_train.shape[1], activation='relu')])
```

```python
# Ajout des couches
for i in range(1, len(sizes)-1):
    Momentum.add(Dense(units=sizes[i], activation='relu'))
    Nag.add(Dense(units=sizes[i], activation='relu'))
    Adagrad.add(Dense(units=sizes[i], activation='relu'))
    Adadelta.add(Dense(units=sizes[i], activation='relu'))
    RMSProp.add(Dense(units=sizes[i], activation='relu'))
    Adam.add(Dense(units=sizes[i], activation='relu'))
```

```python
# Couche de sortie    
Momentum.add(Dense(units=sizes[-1], activation='sigmoid'))
Nag.add(Dense(units=sizes[-1], activation='sigmoid'))
Adagrad.add(Dense(units=sizes[-1], activation='sigmoid'))
Adadelta.add(Dense(units=sizes[-1], activation='sigmoid'))
RMSProp.add(Dense(units=sizes[-1], activation='sigmoid'))
Adam.add(Dense(units=sizes[-1], activation='sigmoid'))
```

```python
layers_sizes = [X_train.shape[1], *sizes]
weigths = []
#print(len(layers_sizes))
```

```python
# Initialisation des poids au même endroit
for i in range(len(layers_sizes)-1):
    weigths.append(np.random.rand(layers_sizes[i], layers_sizes[i+1]))  # la matrice correspondante a la couche ou on se trouve
    weigths.append(np.random.rand(layers_sizes[i+1]))   #la liste du biais
```

```python
Momentum.set_weights(weigths)
Nag.set_weights(weigths)
Adagrad.set_weights(weigths)
Adadelta.set_weights(weigths)
RMSProp.set_weights(weigths)
Adam.set_weights(weigths)

batch_sizes = 20
epochs = 5
```

```python
Momentun_optimizer = tf.keras.optimizers.SGD(
    learning_rate=0.01,
    momentum=0.1,
    nesterov=False,
    name='SGD',
)
# Compiling the ANN
Momentum.compile(optimizer=Momentun_optimizer, loss='binary_crossentropy', metrics=['accuracy'])
# Fitting the ANN to the Training set
Momentum.fit(X_train, y_train, batch_size=batch_sizes, epochs=epochs)
```

```python
Nag_optimizer = tf.keras.optimizers.SGD(
    learning_rate=0.01,
    momentum=0.1,
    nesterov=True,
    name='SGD',
)
# Compiling the ANN
Nag.compile(optimizer=Momentun_optimizer, loss='binary_crossentropy', metrics=['accuracy'])
# Fitting the ANN to the Training set
Nag.fit(X_train, y_train, batch_size=batch_sizes, epochs=epochs)
```

```python
Adagrad_optimizer = tf.keras.optimizers.Adagrad(
    learning_rate=0.01,
    initial_accumulator_value=0.1,
    epsilon=1e-07,
    name='Adagrad',
)
# Compiling the ANN
Adagrad.compile(optimizer=Adagrad_optimizer, loss='binary_crossentropy', metrics=['accuracy'])
# Fitting the ANN to the Training set
Adagrad.fit(X_train, y_train, batch_size=batch_sizes, epochs=epochs)
```

```python
tf.keras.optimizers.Adadelta(
    learning_rate=0.01,
    rho=0.95,
    epsilon=1e-07,
    name='Adadelta',
)
# Compiling the ANN
Adadelta.compile(optimizer=Adagrad_optimizer, loss='binary_crossentropy', metrics=['accuracy'])
# Fitting the ANN to the Training set
Adadelta.fit(X_train, y_train, batch_size=batch_sizes, epochs=epochs)
```

```python
tf.keras.optimizers.RMSprop(
    learning_rate=0.01,
    rho=0.95,
    momentum=0.1,
    epsilon=1e-07,
    centered=False,
    name='RMSprop',
)
# Compiling the ANN
RMSProp.compile(optimizer=Adagrad_optimizer, loss='binary_crossentropy', metrics=['accuracy'])
# Fitting the ANN to the Training set
RMSProp.fit(X_train, y_train, batch_size=batch_sizes, epochs=epochs)
```

```python
tf.keras.optimizers.Adam(
    learning_rate=0.01,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07,
    amsgrad=False,
    name='Adam',
)
# Compiling the ANN
Adam.compile(optimizer=Adagrad_optimizer, loss='binary_crossentropy', metrics=['accuracy'])
# Fitting the ANN to the Training set
Adam.fit(X_train, y_train, batch_size=batch_sizes, epochs=epochs)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
print("Our accuracy is {}%".format(((cm[0][0] + cm[1][1])/57)*100))
```
